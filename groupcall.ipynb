```python
def start_time_hours(x):
    hour_second_start = int(x[0:2])*3600
    miniute_second_start = int(x[3:5])*60
    second_start = float(x[6:])
    return  hour_second_start+ miniute_second_start + second_start
```

```python
df['call_start_time_change'] = df.call_start_time.apply(start_time_hours)
df['call_end_time_change'] = df.call_end_time.apply(start_time_hours)
df['day_diff'] = (df['call_end_day'] - df['call_start_day'])*24*3600
df['call_time'] = df.day_diff + df.call_end_time_change - df.call_start_time_change

del df['call_start_time_change']
del df['call_end_time_change']
del df['day_diff']
```
